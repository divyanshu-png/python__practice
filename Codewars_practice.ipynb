```python
def duplicate_encode(word):
  word=word.lower()
  return "".join(["(" if word.count(char) == 1 else ")" for char in word])

duplicate_encode("recede")
```

```python

def get_sum(a,b):
  return sum(x for x in range(min(a,b),max(a,b)+1))

get_sum(2,-9)
```

```python
def number(bus_stops):
  tot=0
  for[a,b]in bus_stops:
    tot+=a
    tot-=b
  return tot
number([[10,0],[3,5],[5,8]])
```

```python
def solution(number):
  i=sum=0
  for i in range(0,number):
    if i%3==0 or i%5==0:
      sum+=i
  return sum

solution(4)
```

```python
def solution(number):
  i=0
  sum=0
  for i in range(0,number):
    if i//3==0 or i//5==0:
      sum+=i
  return sum

solution(10)
```

```python
# sum of factors of a number
def solution(number):
  return sum(i for i in range(1,number+1) if number%i==0)

solution(20)
```

```python
def binary_array_to_number(arr):
    res=0
    arr[::-1]
    x=len(arr)
    for i in range(0, x):
        res+= arr[i]*(2**(i))
    return res
binary_array_to_number([1, 0, 0, 1])
```
