```python
# Rock paper scissors game
import random
while True:
  if next==False:
    break
  player_choice= input("Enter a choice out of the three")
  computer_choice= random.choice(["rock","paper","scissors"])
  if player_choice==computer_choice:
    print("Draw")
  elif player_choice=="rock":
    if computer_choice=="scissors":
      print("Player wins")
    elif computer_choice=="paper":
      print("Computer wins")
  elif player_choice=="paper":
    if computer_choice=="rock":
      print("Player wins")
    elif computer_choice=="scissors":
      print("Computer wins")
  elif player_choice=="scissors":
    if computer_choice=="paper":
      print("Player wins")
    elif computer_choice=="rock":
      print("Computer wins")
  next= input("Do you wish to play again?(True/False)")
```

 Returning 1 value covers 90% of the cases. But sometimes it really makes sense to return 2 or more values. The Pythonic way to do this is to return a ***tuple*** packing together the values.

```python
def min2(a, b):
    """
    Given 2 ints, returns (True, min_val) if at least
    one is negative and min_val is the minimum value.
    Otherwise returns (False, None)
    """
    if a < 0 or b < 0:
        return (True, min(a, b))
    return (False, None)

min2(-9,5)
```

Another useful data type built into Python is the dictionary (see Mapping Types — dict). Dictionaries are sometimes found in other languages as “associative memories” or “associative arrays”. Unlike sequences, which are indexed by a range of numbers, dictionaries are indexed by keys, which can be any immutable type; strings and numbers can always be keys. Tuples can be used as keys if they contain only strings, numbers, or tuples; if a tuple contains any mutable object either directly or indirectly, it cannot be used as a key. You can’t use lists as keys, since lists can be modified in place using index assignments, slice assignments, or methods like append() and extend().

The main operations on a dictionary are storing a value with some key and extracting the value given the key. It is also possible to delete a key:value pair with del. If you store using a key that is already in use, the old value associated with that key is forgotten.

**SETS**

A set is an unordered collection with no duplicate elements. Basic uses include membership testing and eliminating duplicate entries. Set objects also support mathematical operations like union, intersection, difference, and symmetric difference.

```python
divyanshu={"Divyanshu Singh","Cse-t2", 10515602723,"Data Sceince", 9.00,"IBM"}
DIVYANSHU2={"Divyanshu Singh", "ADGIPS"}
print(divyanshu)
```

because sets are inherently unordered collections. Unlike lists or tuples, which maintain the order of elements as they are inserted, Python sets do not store elements in any specific sequence. Their primary purpose is to store unique elements and perform set operations efficiently, not to preserve order.

```python
print("Divyanshu Singh" in divyanshu) #membership operator
print(divyanshu | DIVYANSHU2 ) #UNION of the two sets with repeating values eliminated
print(divyanshu & DIVYANSHU2) #intersection of the two sets
```

```python
s1={1,2,3,4}
s2={3,4,5,6,7}

print(s1-s2)#Set difference
print(s1^s2) #symmetric difference
print(s1|s2)#Union
print(s1&s2)#Intersection
```
