The map() function applies a specified function to each item in an iterable (like a list or tuple) and returns a new iterable (a map object) with the results of the transformation.

````python
#Filter
def check_even(num):
  return num%2==0

mynums=[1,2,3,4,5,6,7]

list(filter(check_even, mynums))
````


 **Lambda** expression is a small, anonymous function defined without a name. Unlike a regular function created with the def keyword, a lambda is limited to a single expression and is typically used for short, one-time operations.

````python
#common use case of Lambda

#list element's square
my_nums=[1,2,3,4,5]
list(map(lambda num: num**2, my_nums))
````

**Syntax:**

```
lambda arguments: expression
```

How does Python know which "x" are we reffering to when we call various or which "x" am I referring to in my code.

The **LEGB** rule is how Python determines the scope of a variable when it is referenced. The acronym stands for Local, Enclosing, Global, and Built-in, which is the order in which Python searches for a name.


**Local (L)**: Python first checks the current function or class method where the variable name is used. Variables defined here are local to that function and cannot be accessed from outside it.

**Enclosing (E)**: If not found in the local scope, Python looks in the scope of any outer or enclosing functions (applicable in the case of nested functions).

**Global (G)**: If still not found, Python checks the global scope, which is the top level of the script or module. Names defined at this level are accessible from anywhere in the module.

**Built-in (B)**: Finally, Python checks the special built-in scope, which contains names of predefined functions, exceptions, and attributes provided by Python (e.g., print(), len(), str(), open, continue, range(), syntaxerror).

````python
name="this is a global string"

def greet():
  name="Sammy,  This is an enclosed variable"
  def helllo():
    print("hello", name)
  helllo()
````

The **global** keyword in Python is used to indicate that a variable inside a function (or another local scope) refers to the variable in the global scope (the module level), rather than a new local variable

````python
x= 50

def func():
  global x
  print(f"Value of X is {x}")

  x+=100
  print(f'I just locally changed the value of X to {x}')
````

````python
def up_low(s):
  u=[]
  l=[]
  for x in s:
    if x.isupper():
      u.append(x)
    elif x.islower():
      l.append(x)
  print("Number of Uppercase Letters", len(u))
  print("Number of Lowercase letters" , len(l))
````

````python
#palindrome

def palindrome(n):
  if type(n)==str:
    n="".join(n.split())
  return ( list(n)==list(n[::-1]) or n.lower()==n[::-1].lower()) # for list


````

````python
palindrome("helleh")
````

````python
palindrome("Helleh")
````
