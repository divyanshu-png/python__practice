```python
#try-except-finally
try:
  f=open('testfile.txt',"w")
  print("Writing to file")
  f.write("Write a test line")
except TypeError:
  print("There was a type error!")
except OSError:
  print("Hwy you have an OS error")
finally:
  print("This block of code includes something like deleting the file element to free up memory: something that needs to be executed regardless of the error")
```

```python
#try-except-finally
try:
  f=open('testfile.txt',"r")# changed the mode from w to r
  print("Writing to file")
  f.write("Write a test line")
except TypeError:
  print("There was a type error!")
except OSError:
  print("Hey you have an OS error")
finally:
  print("This block of code includes something like deleting the file element to free up memory: something that needs to be executed regardless of the error")
```

```python
def ask_for_integer():
  while True:
    try:
      result= int(input("please provide a number"))
    except:
      print("Whoops! That is not a number")
      print("I'm going to ask you again !")
    else:
      print("Yes thank you!")
      break
    finally:
      print("End of try-except-finally block \nI will always run at the end")


ask_for_integer()
```

Python as a set of style convention rules known as PEP8:   provides conventions to improve code readability and consistency. It mandates 4-space indentation, 79-character line limits, snake_case naming for functions/variables, and consistent whitespace usage. Adhering to PEP 8 makes code easier to read, debug, and maintain.

Naming Conventions:
1. Variables/Functions: lowercase with words separated by underscores (e.g., my_variable).
2. Classes: CamelCase (e.g., MyClass).
3. Constants: ALL_UPPERCASE.
