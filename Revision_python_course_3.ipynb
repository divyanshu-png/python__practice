```python
#@title Warmup project: Card game using OOP
"""LOGIC:
We start of with a deck of 52 cards divinfing into two players eqaully, 26 cards for each player. The two player pick out the topmost/ bottom most card from their deck and lay it on table.
Player that has the higher ranking card wins and adds the card of the other player into their deck. This continues unitl one of the player has the whole deck, who eventually wins the game
If they happen to be of exact same rank, each player set out 3/2 cards each and go on to the next card and now whoever wins this time gets the 3-2 set aside cards of the opponent and the current card as well!"""
"""IMPLEMENTATION:
We can implement this using OOPS :
1. card class
2. deck class
3. player class
4. Game logic """


rank_value_mapping={"Two":2, "Three":3, "Four":4, "Five":5, "Six":6,"Seven":7, "Eight":8, "Nine":9, "Ten":10, "Jack":11, "Queen":12, "King":13, "Ace":14}
values=set(rank_value_mapping.keys())
suits={'Hearts', 'Diamond','Club','Spade'}
```

```python
class card():
    """ each card should have atleast three properties that are suit(diamond, club, space, heart), rank( A,2....J,Q,K), and an integer value """
    def __init__(self, suit, rank):
        self.suit=suit
        self.rank=rank.title()
        self.value=rank_value_mapping[rank]
    def __str__(self):
        return str(self.rank) + " of "+ self.suit
```

```python
from itertools import product
import random
class deck():
    """The class deck needs to have the following features : instantitate a new deck of 52 card objects and hold as a list of card objects
    Secondly create a method to shuffle the deck using the Random Library(irrespective of the length of the deck)
    And lastly, Pop method to pick out one card from the deck as needed to play the game. """

    """We take the set of both the suit and rank of the card"""
    def __init__(self):
        values=set(rank_value_mapping.keys())
        suits={'Hearts', 'Diamond','Club','Spade'}
        list_deck_cards= list(product(values,suits))
        self.all_cards=[]

        for i in list_deck_cards:
            created_card= card(i[1],i[0])
            self.all_cards.append(created_card)
    def shuffle_deck(self):
        random.shuffle(self.all_cards)
    def lift_card_from_deck(self):
        try:
            return self.all_cards.pop(0) # to pop the topmost(teating as a deck)
        except:
            return "Error as there are no more cards to draw"
```

```python
# Player class
class Player():
    """ A player class is used to hold a player's current list of cards.
    A player should be able to do the following:
    1. Add or remove cards from thier hand (list of card objects)
    2. Allow the user to add a single card or multiple cards to the list and also true for the removing cards
    Logic: When you win you want to add cards at the bottom of the deck and when you take out a card you want it to be the topmost card: Thus we have to use the .pop(0) => at index 0
    to add single card we used .append() method and to add a list of cards we use the .extend(<list_name>) method
    """
    def __init__(self, name):
        self.name=name
        self.all_cards=[]
    def remove_one_card(self):
        return self.all_cards.pop(0)
    def add_cards(self, new_cards):
        if type(new_cards)==list:
            self.all_cards.extend(new_cards)# multiple carsd scenario
        else:
            self.all_cards.append(new_cards)#single card scenario
    def __str__(self):
        return f"Player {self.name} has {len(self.all_cards)} cards remaining"
```

```python
# game logic
player_one= Player("One")
player_two= Player("Two")

deck_new= deck()
deck_new.shuffle_deck()
for i in range(0,26):
    player_one.add_cards(deck_new.lift_card_from_deck())
    player_two.add_cards(deck_new.lift_card_from_deck())
game_on= True
round_no=0
while game_on:
    round_no+=1
    print(f"Round {round_no}! FIGHT!")
    if len(player_one.all_cards)==0:
        print(f"{player_two.name} Wins the game")
        game_on=False
        break
    elif len(player_two.all_cards)==0:
        print(f"{player_one.name} Wins the game")
        game_on=False
        break
    #start new round
    player_one_cards_at_table=[]
    player_one_cards_at_table.append(player_one.remove_one_card())
    player_two_cards_at_table=[]
    player_two_cards_at_table.append(player_two.remove_one_card())

    """Game logic checking :
    """
    at_war= True
    while at_war:
        if player_one_cards_at_table[-1].value > player_two_cards_at_table[-1].value:# -1 here as the formation is as of a stack of card and the one at the top is compared
            #player 1 beats player 2 card
            print(f"Player 1 won the Round {round_no}")
            player_one.add_cards(player_two_cards_at_table)
            player_one.add_cards(player_one_cards_at_table) # he keeps his own cards too
            at_war=False
        elif player_two_cards_at_table[-1].value > player_one_cards_at_table[-1].value:
            #player 2 beats player 1
            print(f"Player 2 won the Round {round_no}")
            player_two.add_cards(player_one_cards_at_table)
            player_two.add_cards(player_two_cards_at_table)
            at_war=False
        else :
            print('War!')
            if len(player_one.all_cards)<3:
                print(f"Player one unable to declare war! \n{player_two.name} has won the game")
                game_on= False
                break
            elif len(player_two.all_cards)<3:
                print(f"Player two unable to declare war! \n{player_one.name} has won the game")
                game_on= False
                break
            else:
                for num in range(3):
                    player_one_cards_at_table.append(player_one.remove_one_card())
                    player_two_cards_at_table.append(player_two.remove_one_card())




```

```python
#@title Second Milestone Project


""" BALCKJACK / 21:  simplfifed game between human and computer. We start with a normal deck of cards, (deck object similar to the deck in previous game)

Concept: Human player has a bankroll and places a bet on whether or not they are going to win this set of hands(A list of card objects). The computer dealer starts with one card face up and one card face down. The human player starts wiht both cards face up. Player goes first and the player goal is to :
    Get closer to a total value of 21 than the computer dealer: the total value is the sum of the value of current face up cards the human player has.
Player action: Hit(Recieve another card) or Stay(stop recieving cards)
After player action, if the player is under 21, then dealer hits(revieves another card from the deck) unitl eihter they beat the palyer or the dealer busts( goes over 21 in value )


After the first two cards are dealt to dealer and player, the player is asked if they'd like another card (called 'hitting'), or if they are happy with the cards they have already (called 'staying'). The object is to make the sum of your card values as close to 21, without going over. If we make 21 exactly, we have blackjack, which can't be beat. If we go over 21, we 'bust' and we lose the round. The player is allowed to stop hitting at any point.

Rules forv value of different cards:
The number cards (2 through 10) are worth the number displayed, face cards are worth 10, and an Ace can be worth either 1 or 11(prefferable oout of the two for example: if the sum is 20 value of ace will be used as 1 and if say the total value is 10, the value will  be taken as 11.)

Once our hand is finished, the dealer tries to do the same. The dealer must keep hitting until they get to 17. If they get above 17 without busting, they can stay.


Here is the step-by-step flow of a single player BlackJack game:
0. Bet:  Ask the player the amount of bet they want to place before they are dealt with any cards
1. Deal: The player and computer (dealer) both receive two cards. The player's cards are face-up, while one of the dealer's cards is face-up and one is face-down.
2. Player’s Turn: The player is asked to "Hit" (take another card) or "Stay/Stand" (keep current total).
3. Repetition: The computer repeatedly asks the player if they want to hit or stay until the player chooses to stay or until they bust (total > 21).
4. Dealer’s Turn: Once the player stands, the dealer reveals their face-down card.
5.  Dealer Rules: The computer (dealer) must follow strict rules: it must hit if the total is 16 or less, and stand on 17 or more.
6. Compare: If the player did not bust, the hands(set of cards that are available to the player and the dealer) are compared to determine the winner.
"""
```

```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7,'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}

playing= True
```

```python
class Card():
    def __init__(self, rank, suit):
       self.suit= suit
       self.rank= rank
       self.value=values[rank]
    def __str__(self):
        return self.rank+ " of "+ self.suit
```

```python
class Deck():
    def __init__(self):
       self.deck =[]
       for suit in suits:
        for rank in ranks:
            self.deck.append(Card(rank, suit))
    def __str__(self):
        deck_comp= ''
        for card in self.deck:
            deck_comp+= '\n'+ card.__str__()
        return deck_comp
    def shuffle_deck(self):
        random.shuffle(self.deck)
    def deal(self):
        return self.deck.pop(0)
```

```python
class Hand:

    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces

    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        # tracking the aces
        if card.rank == 'Ace':
            self.aces += 1  # add to self.aces

    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips:

    def __init__(self):
        self.total = 100  # This can be set to a default value or supplied by a user input
        self.bet = 0

    def win_bet(self):
        self.total += self.bet

    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):

    while True:
        try:
            chips.bet = int(input('How many chips would you like to bet? '))
        except ValueError:
            print('Sorry, a bet must be an integer!')
        else:
            if chips.bet > chips.total:
                print("Sorry, your bet can't exceed",chips.total)
            else:
                break
```

```python
def hit(deck,hand):

    hand.add_card(deck.deal())
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop

    while True:
        x = input("Would you like to Hit or Stand? Enter 'h' or 's' ")

        if x[0].lower() == 'h':
            hit(deck,hand)  # hit() function defined above

        elif x[0].lower() == 's':
            print("Player stands. Dealer is playing.")
            playing = False

        else:
            print("Sorry, please try again.")
            continue
        break
```

```python
def show_some(player,dealer):
    print("\nDealer's Hand:")
    print(" <card hidden>")
    print('',dealer.cards[1])
    ##player hands
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    ##


def show_all(player,dealer):
    print("\nDealer's Hand:", *dealer.cards, sep='\n ')
    print("Dealer's Hand =",dealer.value)
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    print("Player's Hand =",player.value)
```

```python
def player_busts(player,dealer,chips):
    print("Player busts!")
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print("Player wins!")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("Dealer busts!")
    chips.win_bet()

def dealer_wins(player,dealer,chips):
    print("Dealer wins!")
    chips.lose_bet()

def push(player,dealer):
    print("Dealer and Player tie! It's a push.")
```

```python
while True:
    # Print an opening statement
    print('Welcome to BlackJack! Get as close to 21 as you can without going over!  \nDealer hits until she reaches 17. Aces count as 1 or 11.')

    # Create & shuffle the deck, deal two cards to each player
    deck = Deck()
    deck.shuffle_deck()

    player_hand = Hand()
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())

    dealer_hand = Hand()
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())

    # Set up the Player's chips
    player_chips = Chips()  # remember the default value is 100

    # Prompt the Player for their bet
    take_bet(player_chips)

    # Show cards (but keep one dealer card hidden)
    show_some(player_hand,dealer_hand)

    while playing:  # recall this variable from our hit_or_stand function

        # Prompt for Player to Hit or Stand
        hit_or_stand(deck,player_hand)

        # Show cards (but keep one dealer card hidden)
        show_some(player_hand,dealer_hand)

        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chips)
            break


    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
    if player_hand.value <= 21:

        while dealer_hand.value < 17:
            hit(deck,dealer_hand)

        # Show all cards
        show_all(player_hand,dealer_hand)

        # Run different winning scenarios
        if dealer_hand.value > 21:
            dealer_busts(player_hand,dealer_hand,player_chips)

        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand,dealer_hand,player_chips)

        elif dealer_hand.value < player_hand.value:
            player_wins(player_hand,dealer_hand,player_chips)

        else:
            push(player_hand,dealer_hand)

    # Inform Player of their chips total
    print("\nPlayer's winnings stand at",player_chips.total)

    # Ask to play again
    new_game = input("Would you like to play another hand? Enter 'y' or 'n' ")

    if new_game[0].lower()=='y':
        playing=True
        continue
    else:
        print("Thanks for playing!")
        break
```

In Python, an iterator is an object used to traverse through a collection of data (like lists, tuples, or dictionaries) one element at a time. It is the engine that powers for loops, comprehensions, and iterable unpacking.

1.  The Iterator Protocol
To be considered an iterator, an object must implement two specific "dunder" (double underscore) methods, collectively known as the iterator protocol:

__iter__(): Returns the iterator object itself. This is required so that both containers and iterators can be used in for loops.

__next__(): Returns the next value from the data stream. If no more items are available, it must raise a StopIteration exception.

2. Iterator vs. Iterable
These terms are often confused but have distinct roles:
Iterable: Any object you can loop over (e.g., list, str, dict). These objects have an __iter__() method that returns a new iterator.
Iterator: The actual object that performs the traversal. It keeps track of its current state and can only be used once; once exhausted, it cannot be reset.

3. Rule of Thumb: Every iterator is an iterable, but not every iterable is an iterator.

The exact decorator syntax used here is the following:

 `@ < decorator_function_name >`

 Thus, Decorator is a function that takes another function as an argument, allowing you to extend its behavior without modifying its source code.
 1. Decorators are applied using the @decorator_name syntax above a function definition.
 2. Use *args and **kwargs in the wrapper to ensure compatibility with any function arguments.
 3. Inside the decorator, an inner function (often called wrapper) is defined to execute code before and after the original function.

 This capability relies on functions being first-class objects, meaning they can be passed as arguments, returned, or assigned to variables.

We won't be using decorators that much. Common applications include logging, timing, authentication in web frameworks (e.g., Flask/Django), and caching (@functools.lru_cache).

Frequently used built-ins include @property, @staticmethod, and @classmethod.

```python
#example from FreeCodeCamp

import functools

def decoratorFunction(func):
    @functools.wraps(func)
    def wrapperFunction(*args, **kwargs):
        print('Wrapper executed before {}'.format(func.__name__))
        return func(*args, **kwargs)
    return wrapperFunction

@decoratorFunction
def display():
    print('The display function was called')

@decoratorFunction
def display_info(name, age):
    print('display_info was called with ({}, {})'.format(name, age))

display_info('Kalam', 83)
display()
```

```python
 #@title Generators in Python
"""Generator functions allow us to write a function that can send back a value and then later pick it up where it left off.
 A generator is a special type of function that returns an iterator object, allowing you to produce a sequence of values one at a time. Unlike regular functions that compute an entire result and return it at once, generators use lazy evaluation, producing values only when requested."""
```

When a generator function is compileed they become an object that supports an iteration protocol. That means when they are called in your code they dont actually return a value and then exit. Instead they will automatically suspend and resume thier execution and their state around the last point of value generation.

```python
#yeild keyword statement

def count_to(n):
    numbers=[]
    count=1
    while count<=n:
        numbers.append(count)
        count+=1
    return numbers
number=int(input("Enter a number "))

for n in count_to(number): # using a function as an iterable item
    print(n)
```

```python
#the changes to be made for this to work:

def count_to(n):
    count=1
    while count<=n:
        yield count
        count+=1
number=int(input("Enter a number "))

for n in count_to(number): # using a function as an iterable item
#here in this for loop, the count_to(number) returns an object that has an value uodated each time after the
    print(n)
#it will take time but we can see the results and the computer will not run into an error or dead state like witht the previous case
```

The main benefit of using yield and a generator here (especially for very large numbers) is memory efficiency. Instead of building a potentially massive list of all numbers in memory and then returning it (like the previous example in your notebook did), the generator produces numbers one at a time, only when they are needed by the loop. This prevents your program from running out of memory for large inputs.

```python
#the changes to be made for this to work:

def count_to(n):
    count=1
    while count<=n:
        yield count
        count+=1
number=int(input("Enter a number "))

for n in count_to(number): # using a function as an iterable item
#here in this for loop, the count_to(number) returns an object that has an value uodated each time after the
    print(n)
#it will take time but we can see the results and the computer will not run into an error or dead state like witht the previous case
```

```python
from os import read
def read_file(file_path):
    with open(file_path) as file:
        for line in file:
            yield line.strip()

file_path="/content/My_Book.txt"
for line in read_file(file_path):
    print(line)
#even though it is a small file, for larger files a generator works the best
```

```python
#create a generator that yields 'n' random numbers between a low and high number that are inputs

import random
def rand_num(low, high, n):
    ite=0
    for i in range(low, high):
        while ite<=n:
            yield random.randint(low, high)
            ite+=1
```
