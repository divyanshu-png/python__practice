Regular Expression Pattern:

1. \d: digits
2. \w: Alpha-numeric(Characters that are either letters or numbers and also include underscore(_))
3. \s: whitespace
4. \D: a non digit
5: \W: Non- alpha-numeric
6. \S: Non-whitespace


````python
text1 = "Python310"
text2 = "Python 3.10"

print(text1.isalnum())  # True (letters and numbers only)
print(text2.isalnum())  # False (contains spaces and a period)
````

Technical Types of White Space
Python recognizes several characters as white space, primarily defined in the string.whitespace constant:

0. Space (' '): The standard ASCII space character (0x20).
1. Horizontal Tab (\t): Used for indentation; typically represents multiple spaces in text editors.
2. Newline (\n): Also called line feed; ends a physical line of code.
3. Carriage Return (\r): Part of Windows-style line endings (\r\n).
4. Form Feed (\f): A page break character, sometimes used in older documentation tools to separate sections.
5. Vertical Tab (\v): A vertical spacing character, rarely used in modern code.

Quantifiers that can be used in front of the RE pattern that help us identify repeating pattern object:
1. +: Occurs one or more time
2. {n}: Occurs exactly n times
3. {n,m}: Occurs n to m times
4. {n,}: Occurs n or more time(minimum frequency is 3)
5. *: Occurs zero or more time
6. ?: Once or none occurance

````python
#@title Timing your code

#For multiple solutions for a single task, it helps in figuring out the most time efficient approach. An easy way to do so is to time your code's performance


def func_one(n):
    return [str(num) for num in range(n)]

def func_two(n):
    return list(map(str, range(n)))
````

````python
func_one(10)
````

````python
func_two(10)
````

````python
import time
start_time= time.time()
func_one(100000)
end_time=time.time()
print(end_time-start_time)
````

````python
import time
start_time= time.time()
func_two(100000)
end_time=time.time()
print(end_time-start_time)
````

````python
#this gives us the idea that fun_one is faster in some cases and takes a little bit less time than func_two
#in other use cases func_two performs better than func_one
# however for small use case the differnece becomes negligible and reduces even less than 1/10th of a second

import time
start_time= time.time()
func_one(10)
end_time=time.time()
print(end_time-start_time)
````

````python
import time
start_time= time.time()
func_two(10)
end_time=time.time()
print(end_time-start_time)
````

````python
#TIMEIT MODULE
import timeit

# func_one(10)
stmt="""
func_one(1000)
"""
setup="""
def func_one(n):
    return [str(num) for num in range(n)]
"""

timeit.timeit(stmt, setup, number=100000)
````

````python
import timeit

# func_two(10)
stmt2="""
func_two(1000)
"""
setup2="""
def func_two(n):
    return list(map(str, range(n)))
"""

timeit.timeit(stmt2, setup2, number=100000)
#thus here we see that func_one is the more efficient function
````

````python
#iteration through each page
result=[]
for i in range(1,51):
    scrap_url= f"https://books.toscrape.com/catalogue/page-{i}.html"
    # print(base_url)
    res= requests.get(scrap_url)
    soup= bs4.BeautifulSoup(res.text, 'lxml')
    products= soup.select('.product_pod')
    for product in products:
        #alternativeley can use
        # if  'star-rating Two' in str(product):
        #or
        # if not [] == product.select('.star-rating.Two'):
        if len(product.select('.star-rating.Two'))!=0:
            book_title = product.select('a')[1]['title']
            result.append(book_title)
result
````

````python
two_star_titles = []
base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
for n in range(1,51):

    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)

    soup = bs4.BeautifulSoup(res.text,"lxml")
    books = soup.select(".product_pod")

    for book in books:
        if len(book.select('.star-rating.Two')) != 0:
            two_star_titles.append(book.select('a')[1]['title'])
````

Another way described by the tutor:

````python
url = 'http://quotes.toscrape.com/page/'
page_still_valid = True
authors = set()
page = 1

while page_still_valid:

    # Concatenate to get new page URL
    page_url = url+str(page)

    # Obtain Request
    res = requests.get(page_url)

    # Check to see if we're on the last page
    if "No quotes found!" in res.text:
        break

    # Turn into Soup
    soup = bs4.BeautifulSoup(res.text,'lxml')

    # Add Authors to our set
    for name in soup.select(".author"):
        authors.add(name.text)

    # Go to Next Page
    page += 1
````
