```python
#@title Regular Expression Library

# Regular expressions or regex for short allow us to search for general patterns in text data. For example, a simple email format can be : user@email.com

# we know in this case we're looking for a patttern:
#               'text'+@+'text'+'.com'
```

The RE library alllows us to create specialized patterns strings and then search for matches within text.

The primary skill set for regex is understanding the special syntax for these pattern strings.

example of the regex pattern of a phone number within a very big file:

`r"(\d\d\d)-\d\d\d-\d\d\d\d"`

This \d is a placeholder and in this case, stands for digit.

similar to the one mentioned above we can dedscribe it as:
`r"(\d{3})-\d{3}-\d{4}"`

```python
text= "the agent's phone number is 408-995-8541"
"phone" in text
```

```python
import re
pattern ='phone'
re.search(pattern, text)
```

```python
pattern_new= "Not in text"
result =re.search(pattern_new, text) #nothing is returned so there is no matching substring
```

```python
type(result)
```

```python
match_text=re.search(pattern, text)
match_text
```

```python
text_new = 'My phone once, my phone twice'
pattern='phone'
match_text=re.search(pattern, text_new)
match_text
#returns the first match it encounters
```

```python
match_text_2 = re.findall(pattern, text_new)
print(match_text_2)
```

```python
type(match_text_2)
```

```python
for match_text in re.finditer(pattern, text_new):
    print(match_text)
    #returns the match objects
```

```python
for match_text in re.finditer(pattern, text_new):
    print(match_text.group())
    #returns the actual text that you were searching for
```
