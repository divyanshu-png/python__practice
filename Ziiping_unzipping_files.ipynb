```python
#@title Zipping and unzipping files 

f=open('file1.txt', 'w+')
f.write("oNe fILE ")
f.close()
```

```python

f=open('file2.txt', 'w+')
f.write("tWO fILE ")
f.close()
```

```python
import zipfile
comp_file= zipfile.ZipFile('comp_file.zip', 'w')
```

```python
#zipping file1 and file2 in the comp_file
comp_file.write('file1.txt', compress_type= zipfile.ZIP_DEFLATED)
comp_file.write('file2.txt', compress_type=zipfile.ZIP_DEFLATED)

comp_file.close()
```

```python
#extracting items from the zip filter
zip_obj= zipfile.ZipFile('comp_file.zip','r')
zip_obj.extractall("extracted_content")
```

```python
#zip an entire folder or archive(zip) and unzip the entire folder(directory)
```

```python
import shutil
#pointing out an directory that we want to turn into a zip file 
#turning extracted_content/ into a zipped folder

dir_to_zip= r'D:\New folder (3)\PYTHON\extracted_content'
```

```python
output_file= 'example'
shutil.make_archive(output_file, format='zip', root_dir=dir_to_zip)
```

```python
shutil.unpack_archive('example.zip', 'final_unzip', 'zip')
```
