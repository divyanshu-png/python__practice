```python
#@title Second Milestone Project

%%file python_blackjack_game.py

""" BALCKJACK / 21:  simplfifed game between human and computer. We start with a normal deck of cards, (deck object similar to the deck in previous game)

Concept: Human player has a bankroll and places a bet on whether or not they are going to win this set of hands(A list of card objects). The computer dealer starts with one card face up and one card face down. The human player starts wiht both cards face up. Player goes first and the player goal is to :
    Get closer to a total value of 21 than the computer dealer: the total value is the sum of the value of current face up cards the human player has.
Player action: Hit(Recieve another card) or Stay(stop recieving cards)
After player action, if the player is under 21, then dealer hits(revieves another card from the deck) unitl eihter they beat the palyer or the dealer busts( goes over 21 in value )


After the first two cards are dealt to dealer and player, the player is asked if they'd like another card (called 'hitting'), or if they are happy with the cards they have already (called 'staying'). The object is to make the sum of your card values as close to 21, without going over. If we make 21 exactly, we have blackjack, which can't be beat. If we go over 21, we 'bust' and we lose the round. The player is allowed to stop hitting at any point.

Rules forv value of different cards:
The number cards (2 through 10) are worth the number displayed, face cards are worth 10, and an Ace can be worth either 1 or 11(prefferable oout of the two for example: if the sum is 20 value of ace will be used as 1 and if say the total value is 10, the value will  be taken as 11.)

Once our hand is finished, the dealer tries to do the same. The dealer must keep hitting until they get to 17. If they get above 17 without busting, they can stay.


Here is the step-by-step flow of a single player BlackJack game:
0. Bet:  Ask the player the amount of bet they want to place before they are dealt with any cards
1. Deal: The player and computer (dealer) both receive two cards. The player's cards are face-up, while one of the dealer's cards is face-up and one is face-down.
2. Player’s Turn: The player is asked to "Hit" (take another card) or "Stay/Stand" (keep current total).
3. Repetition: The computer repeatedly asks the player if they want to hit or stay until the player chooses to stay or until they bust (total > 21).
4. Dealer’s Turn: Once the player stands, the dealer reveals their face-down card.
5.  Dealer Rules: The computer (dealer) must follow strict rules: it must hit if the total is 16 or less, and stand on 17 or more.
6. Compare: If the player did not bust, the hands(set of cards that are available to the player and the dealer) are compared to determine the winner.
"""


rank_value_mapping={"Two":2, "Three":3, "Four":4, "Five":5, "Six":6,"Seven":7, "Eight":8, "Nine":9, "Ten":10, "Jack":10, "Queen":10, "King":10, "Ace":11}
values=set(rank_value_mapping.keys())
suits={'Hearts', 'Diamond','Club','Spade'}

from itertools import product
import random
class Deck():
    """The class deck needs to have the following features : instantitate a new deck of 52 card objects and hold as a list of card objects
    Secondly create a method to shuffle the deck using the Random Library(irrespective of the length of the deck)
    And lastly, Pop method to pick out one card from the deck as needed to play the game. """

    """We take the set of both the suit and rank of the card"""
    def __init__(self):
        values=set(rank_value_mapping.keys())
        suits={'Hearts', 'Diamond','Club','Spade'}
        list_deck_cards= list(product(values,suits))
        self.all_cards=[]

        for i in list_deck_cards:
            created_card= card(i[1],i[0])
            self.all_cards.append(created_card)
    def shuffle_deck(self):
        random.shuffle(self.all_cards)
    def lift_card_from_deck(self):
        # the function call returns a card that can then be appended to the player's or the dealer's hand
        try:
            return self.all_cards.pop(0) # to pop the topmost(teating as a deck)
        except:
            return "Error as there are no more cards to draw"

class card():
    """ each card should have atleast three properties that are suit(diamond, club, space, heart), rank( A,2....J,Q,K), and an integer value """
    def __init__(self, suit, rank):
        self.suit=suit
        self.rank=rank.title()
        if self.rank == "Ace":
            self.value = 11
        else:
            self.value= rank_value_mapping[rank]


    def __str__(self):
        return str(self.rank) + " of "+ self.suit

class Player:
    def __init__(self, name, total_value_at_player=0):
       self.name=name
       self.hands=[]
       self.total_value_at_player= total_value_at_player # This will be updated by calculate_at_hand

    def calculate_at_hand(self):
        value = 0
        num_aces = 0
        for card_obj in self.hands:
            if card_obj.rank == "Ace":
                num_aces += 1
                value += 11 # Assume 11 initially
            else:
                value += card_obj.value

        # Adjust for Aces if busted
        while value > 21 and num_aces > 0:
            value -= 10 # Change one Ace from 11 to 1
            num_aces -= 1
        return value

    def hit(self, new_card):
        self.hands.append(new_card)
        self.total_value_at_player = self.calculate_at_hand() # Update total using the new calculation logic

    def skip(self):
        pass


class Dealer:
    def __init__(self,total_value_at_dealer=0):
        self.hands=[]
        self.total_value_at_dealer= total_value_at_dealer # This will be updated by calculate_at_hand

    def calculate_at_hand(self):
        value = 0
        num_aces = 0
        for card_obj in self.hands:
            if card_obj.rank == "Ace":
                num_aces += 1
                value += 11 # Assume 11 initially
            else:
                value += card_obj.value

        # Adjust for Aces if busted
        while value > 21 and num_aces > 0:
            value -= 10 # Change one Ace from 11 to 1
            num_aces -= 1
        return value

    def hit(self,new_card):
        self.hands.append(new_card)
        self.total_value_at_dealer = self.calculate_at_hand() # Update total using the new calculation logic

###############################
#Game logic begins
playing=True
while playing:
    availabe_chips_at_player = 5000000
    availabe_chips_at_dealer= 10000000000
    print(f"The dealer has a total of {availabe_chips_at_dealer} chips")
    total_money_on_bet= int(input(f"Enter the total chips you want to bet:(out of {availabe_chips_at_player}) "))
    while total_money_on_bet < availabe_chips_at_player:
        new_deck= Deck()
        new_deck.shuffle_deck()
        player=Player("Divyanshu",0)
        dealer= Dealer(0)
        # give two cards to each player and dealer and append in the hands list of the player and dealer
        for i in range(0,2):
            player.hands.append(new_deck.lift_card_from_deck())
            dealer.hands.append(new_deck.lift_card_from_deck())

        # Update total values after initial deal
        player.total_value_at_player = player.calculate_at_hand()
        dealer.total_value_at_dealer = dealer.calculate_at_hand()

        print("\nPlayer's Hand:")
        for card_obj in player.hands:
            print(card_obj)
        print(f"Player's Total Value: {player.total_value_at_player}")

        while player.calculate_at_hand()<=21:
            choice=input("Do you wish to Hit or Stay ")
            if choice.lower()=='hit':
                player.hit(new_deck.lift_card_from_deck())
                print("\nPlayer's Hand:")
                for card_obj in player.hands:
                    print(card_obj)
                print(f"Player's Total Value: {player.total_value_at_player}")
            elif choice.lower()=='stay':
                player.skip()
                print("\nPlayer's Hand:")
                for card_obj in player.hands:
                    print(card_obj)
                print(f"Player's Total Value: {player.total_value_at_player}")
                break
        while dealer.calculate_at_hand()<17:
            dealer.hit(new_deck.lift_card_from_deck())

        if player.total_value_at_player>21:
            print(f"Player Busted!Dealer wins with {dealer.total_value_at_dealer}!")
            availabe_chips_at_dealer+=total_money_on_bet
            total_money_on_bet=0

        elif dealer.total_value_at_dealer>21:
            print(f"Dealer Busted! Player wins with {player.total_value_at_player}!")
            total_money_on_bet*=2

        elif player.total_value_at_player<21 and dealer.total_value_at_dealer<21 :
            if player.total_value_at_player > dealer.total_value_at_dealer:
                print("Player won!")
                print(f"The total value at dealer was {dealer.total_value_at_dealer}")
                total_money_on_bet*=2
            elif player.total_value_at_player < dealer.total_value_at_dealer:
                print(f"Total value at dealer was {dealer.total_value_at_dealer}")
                print("Player lost!")
                availabe_chips_at_dealer+=total_money_on_bet
                total_money_on_bet=0

        else:
            print(f"Its a push! Player has {player.total_value_at_player} and Dealer has {dealer.total_value_at_dealer}")

        playing= input("Do you wish to continue? y/n")
        if playing.lower()=='n':
            playing=False
            print(f"Player walks away with {total_money_on_bet} chips")
            break
```
